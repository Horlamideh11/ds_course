```python
import pandas as pd 
```

```python
df = pd.read_csv("glassdoor_jobs.csv")
df = df[df['Salary Estimate'] != '-1']
df.tail(50)
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary.head(50)
```

```python
remove_kd_salary = salary.apply(lambda x: x.replace('$','').replace('K',''))
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df['hr_min'] = remove_kd_salary.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['min_salary'] = df['hr_min'].apply(lambda x: int(x.split('-')[0]))
```

```python
df['max_salary'] = df['hr_min'].apply(lambda x: int(x.split('-')[1]))
```

```python
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
#check company name
df['Company text'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3],axis=1)
```

```python
df['Company text'] = df['Company text'].apply(lambda x: x.replace('\n',''))
```

```python
# take location for state
df['Job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
#how many jobs are available in each state
df['Job_state'].value_counts()
```

```python
df.columns
```

```python
#check to see if the job position is at the headquarters
df['same_state'] = df.apply(lambda x: 1 if x.Headquarters == x.Location else 0, axis = 1)
```

```python
# age of company
df['age_of_company'] = df['Founded'].apply(lambda x: x if x < 0 else 2020-x)
```

```python
#python requirement
df['python_req'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['python_req'].value_counts()
```

```python
df['r_req'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['r_req'].value_counts()
```

```python
df['apache_req'] = df['Job Description'].apply(lambda x: 1 if 'apache' in x.lower() or 'apache spark' in x.lower() else 0)
df['apache_req'].value_counts()
```

```python
df['aws_req'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['aws_req'].value_counts()
```

```python
df['excel_req'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['excel_req'].value_counts()
```

```python
df_out = df.drop(['Unnamed: 0'],axis=1)
```

```python
df_out.to_csv('salary_data_cleaned.csv',index = False)
```
